```python
# PROYECTO 01: Pre-procesamiento de datos para modelos de lenguaje (NLP)
# Objetivo: Limpiar texto crudo para eliminar ruido y estandarizar el contenido.

import re

# 1. Datos de entrada (Simulando comentarios de clientes de "Rancho Burger")
comentarios_sucios = [
    "  LA hamburguesa estaba... EXCELENTE!!!  ",
    "Pésimo servicio,,, muy tardado el pedido.  ",
    "Me encantó el sabor de las Alitas!!!! 10/10",
    "no me gusto la salsa :( , estaba muy picosa",
    "  EL ENVÍO LLEGÓ RÁPIDO   "
]

def limpiar_texto(texto):
    # Convertir todo a minúsculas (Estandarización)
    texto = texto.lower()
    # Eliminar caracteres especiales y puntuación excesiva
    texto = re.sub(r'[^\w\s]', '', texto)
    # Eliminar espacios extra al inicio y al final
    texto = texto.strip()
    return texto

# 2. Ejecutar la limpieza
comentarios_limpios = [limpiar_texto(c) for c in comentarios_sucios]

# 3. Mostrar resultados
print("--- DATOS ORIGINALES ---")
print(comentarios_sucios)
print("\n--- DATOS LIMPIOS PARA IA ---")
print(comentarios_limpios)

# 4. Bonus: Etiquetado simple (Análisis de Sentimiento Básico)
def etiquetar_comentario(texto):
    positivas = ['excelente', 'encanto', 'sabor', 'rapido', '1010']
    if any(palabra in texto for palabra in positivas):
        return "POSITIVO"
    return "NEUTRAL/NEGATIVO"

print("\n--- RESULTADO DEL ETIQUETADO ---")
for c in comentarios_limpios:
    print(f"Texto: {c} | Etiqueta: {etiquetar_comentario(c)}")
```
